```python
from geopy.geocoders import Nominatim

def get_latitude_longitude(location_name):
    # Initialize Nominatim geocoder
    geolocator = Nominatim(user_agent="geoapiExercises")

    # Get location information
    location = geolocator.geocode(location_name)

    if location:
        latitude = location.latitude
        longitude = location.longitude
        return latitude, longitude
    else:
        print(f"Location '{location_name}' not found.")
        return None, None

lc =[]
for i in range(len(top_cheap)):
    print(get_latitude_longitude(top_cheap['location'].iloc[i] + ", Lahore, Pakistan"))
    ## Store these values:
    lc.append(get_latitude_longitude(top_cheap['location'].iloc[i] + ", Lahore, Pakistan"))


```

```python
import folium
# Define the city name
city_name = 'Karachi'  # Change this to the desired city
city_df = df[df['city'] == city_name]
# Calculate average price for each location
avg_prices = city_df.groupby('location')['price'].mean().reset_index()

# Sort locations by price to identify top expensive and cheap locations
top_expensive = avg_prices.nlargest(10, 'price')
top_cheap = avg_prices.nsmallest(10, 'price')

# Get the top expensive and cheap locations
top_expensive_locations = top_expensive['location'].tolist()
top_cheap_locations = top_cheap['location'].tolist()

# Filter the dataset for the specified city and top expensive locations
expensive_df = df[(df['city'] == city_name) & (df['location'].isin(top_expensive_locations))]

# Filter the dataset for the specified city and top cheap locations
cheap_df = df[(df['city'] == city_name) & (df['location'].isin(top_cheap_locations))]

# Create a Folium map centered at the mean latitude and longitude of the city
m = folium.Map(location=[expensive_df['latitude'].mean(), expensive_df['longitude'].mean()], zoom_start=12)

# Add markers for expensive locations
for idx, row in expensive_df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['location'] + f" (Expensive: ${row['price']:,.2f})",
        icon=folium.Icon(color='red')  # Red markers for expensive locations

    ).add_to(m)

# Add markers for cheap locations
for idx, row in cheap_df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['location'] + f" (Cheap: ${row['price']:,.2f})",
        icon=folium.Icon(color='green')  # Green markers for cheap locations

    ).add_to(m)

# Display the map
m

```
